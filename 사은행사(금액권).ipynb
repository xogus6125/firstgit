```python
import pandas as pd
import numpy as np
from numpy import int64
# dfa, dfb 숫자에 모든공백 제거(완료.)
# dropna는 NaN 데이터 삭제해주는것. axis는 데이터 프레임의 방향임. 0은 열방향, 1은 행방향
# 시작번호 열을 만들었음 이름은 dfa
dfs = pd.read_excel("C:/Users/silve/OneDrive/바탕 화면/금액권 입출고대장.xlsx")
dfa = dfs['Unnamed: 1'][3:].dropna(axis=0)
dfa.reset_index(drop=True, inplace=True)
dfa = dfa.str.replace(' ', '')


# 끝번호 열 만듬 이름은 dfb
# dfb는 dfb에서 문자열 공백을 없애서 넣는다.
dfb = dfs['Unnamed: 2'][3:].dropna(axis=0)
dfb.reset_index(drop=True, inplace=True)
dfb = dfb.str.replace(' ', '')


# 수량 열 만듬 이름은 dfc 이걸로 수량 int값의 len만큼 시리얼번호 데이터프레임 해당 값을 비교하고 틀리면 개수와 해당 번호 반환.
dfc = dfs['Unnamed: 3'][3:].dropna(axis=0)
dfc.reset_index(drop=True, inplace=True)

# 근무자 열 만듬. 이름은 dfd 이걸로 틀린거 찾으면 이름까지 반환해보기. ex: 1~100 틀린값을 찾으면 그 값의 제일 첫번째가 인덱스 몇번인지. 근무자이름 나타내기.
dfd = dfs['Unnamed: 4'][3:].dropna(axis=0)
dfd.reset_index(drop=True, inplace=True)

# 반납 시작번호 열 만듬 이름은 dfe 이걸로 만약에 해당열에 dfa나 dfb가 값이 있으면 무시하고 반납만 진행하는걸로.
# 반납 떠올때 dropna하면 안되고 기준을 dfa의 index값으로 잡아서 해당 수량만큼 떠오기.(그냥 NaN 값을 없애는 걸로 방향을 잡았음. 샾이나 0으로 채울거임.)
dfe = dfs['Unnamed: 6'][3:].dropna(axis=0)
dfe.reset_index(drop=True, inplace=True)
dfe = dfe.str.replace(' ', '')

# 반납 끝번호 열 만듬.
dff = dfs['Unnamed: 7'][3:].dropna(axis=0)
dff.reset_index(drop=True, inplace=True)
dff = dff.str.replace(' ', '')

# 반납 수량 만듬. 이름은 dfg
dfg = dfs['Unnamed: 8'][3:].dropna(axis=0)
dfg.reset_index(drop=True, inplace=True)

# 데이터 리스트 만들어서 열 데이터 넣음.
data = []
data.append(dfa)
data.append(dfb)
data.append(dfc)
data.append(dfd)
data.append(dfe)
data.append(dff)
data.append(dfg)

# 데이터 프레임을 만들었습니다. concat은 더해주는 함수임. axis는 방향이고 데이터를 오른쪽으로 정렬한다는 뜻.
df = pd.concat([dfa, dfb, dfc, dfd, dfe, dff, dfg], axis = 1)
df.rename(columns={"Unnamed: 1": "시작번호", "Unnamed: 2": "끝번호", "Unnamed: 3": "수량", "Unnamed: 4": "근무자", "Unnamed: 6": "반납 시작번호", "Unnamed: 7": "반납 끝번호", "Unnamed: 8": "반납수량"}, inplace=True)

C = [] # C 리스트에는 끝번호 + 1 한 값 넣어서 다시 붙여넣기할거임.
D = [] # 이건 아직 쓸모없음.
E = []
df_sharp = df['반납 시작번호'] == '#'
Spend = df[df_sharp] # 반납없이 끝까지 금액권 나간거
Spend = Spend[['시작번호','끝번호']] # 이거는 금액권 다 안나간 수량을 뺀 값.
for i in Spend['끝번호']:
    C.append(int(i) + int(1))
Spend['끝번호'] = C # Spend['끝번호'] = C # 끝번호에서 + 1한 값을 넣었음.
Spend.rename(columns = {'시작번호' : 'start', '끝번호' : 'end'}, inplace = True)
Spend['end'] = Spend['end'].astype(str)


B1 = df['반납 시작번호'] != '#'
B1 = df[B1]
B1 = B1[['시작번호', '반납 시작번호']]
B1.rename(columns = {'시작번호' : 'start', '반납 시작번호' : 'end'}, inplace = True)


alldf = pd.merge(Spend, B1, how='outer')
alldf['end'] = alldf['end'].apply(int)
alldf['start'] = alldf['start'].apply(int)
alldf['count'] = alldf['end'] - alldf['start']

last_df = alldf.sort_values(by = 'start', ascending = True)
last_df = last_df.reset_index(drop=True)

last_count = last_df['count'].sum()
print("금액권 전부 회수 후(반품 x) 기입한 총 개수는 " + str(last_count) + " 입니다.")
print("배열 생성중...")
for i in last_df['start']:
    D.append(i)

for i in last_df['end']:
    E.append(i)
def fill_in_range(D, E):
    if len(D) == len(E):
        new_list = []
        for x,y in zip(D, E):
            new_list.extend(np.arange(x,y))
        return pd.DataFrame(new_list)
    else:
        return 'Lists are not the same size'

fillist = fill_in_range(D, E)
fillist.rename(columns = {0 : 'serial'}, inplace=True)
fillist.astype(int)

lastdata = last_data['시리얼번호']
lastdata = lastdata.reset_index(drop=True)
lastdata = lastdata.apply(int64)
lastdata = pd.DataFrame(lastdata)
# lastdata, fillist 두개 series로 만들어주어야함.
lastdata = lastdata.iloc[:,0]
fillist = fillist.iloc[:,0]
lastone = fillist.iloc[:-1] != lastdata.iloc[:-1]
csvpath = r'C:\Users\silve\OneDrive\바탕 화면\kakaoplus\lastdata(final).csv'
dfdf = pd.concat([fillist, lastdata], axis=0)
dfdf = dfdf.drop_duplicates(keep = False)
dfdf = dfdf.sort_values(ascending = True)
print(str(UserInput) + " 날짜의 로스(숫자 이상)는 " + str(len(dfdf)) + "장입니다."
"""
lastdata(final).csv 파일을 확인하세요
내역은 이렇습니다.
""")
dfdf


# lastdata.to_csv(csvpath)


# lastone = lastdata == fillist
# lastdata = lastdata[lastone]
# lastdata
# lastdata.to_csv(csvfile)
# last_F = pd.DataFrame(lastdata, fillist)
# last_F.to_csv(csvfile)
```
